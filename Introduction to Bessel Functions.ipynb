```python
from scipy.special import jv
from scipy.stats import norm, cauchy
import math
import numpy as np
from matplotlib import pyplot as plt
```

## Bessel Function
$\displaystyle \Large J_0 = \sum^{\infty}_{k=0} (-1)^k \frac{x^{2k}}{2^{2k}(k!)^2} \\$ 
$ \Large = 1 - \frac{1}{2^2}x^2 + \frac{1}{2^4 (2!)^2}x^4 - \frac{1}{2^6 (3!)^2}x^6+...$

```python
scipy.special.jv(0, 1.)
```

```python
######################
###### Constants #####
######################

k=0

######################
###### Plotting ######
######################


# Generate x values, values are in Megahertz
x = np.linspace(-25, 25, 10000) 

# Number of rows and cols in fig
ax_rows, ax_cols = 1, 1
fig, ax = plt.subplots(ax_rows, ax_cols, figsize=(10, 10))

# Plot the normal distribution
ax.set_title('Intensity against Position', fontsize=18)
ax.plot(x, scipy.special.jv(k, x), label=f'k={k}')
ax.plot(x, norm.pdf(x)*2.5, ls='--', label='Normal dist', alpha=0.5)
ax.plot(x, cauchy.pdf(x)*3.1, ls='--', label='Lorentz dist', alpha=0.5)

ax.text(2, 1,"Bright central region", fontsize=16)
ax.text(6, -0.4,"First dark ring (x=3.83)", fontsize=16)
ax.text(6, 0.33,"First bright ring (x=7.02)", fontsize=16)
ax.tick_params(labelsize=18)
ax.legend(loc="upper left", fontsize=18)
ax.grid(alpha=0.5)
plt.show()
#plt.savefig('Bessel Function of the first kind of order one.png', transparent=True)
```
